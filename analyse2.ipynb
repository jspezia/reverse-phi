```python
import numpy as np
import pylab
from scipy.optimize import curve_fit
 
def sigmoid(c, c0, k):
    y = 1 / (1 + np.exp(-k*(c-c0)))
    return y

def psycho_curve(results, com='toto'):
    N_bin = 10
    mean, slope = [], []
    cdata, ydata = results[:, 2], .5*results[:, 0]+.5
    pylab.plot(cdata, ydata, 'o')
    popt, pcov = curve_fit(sigmoid, cdata, ydata)
    mean.append(popt[0])
    slope.append(popt[1])
    c = np.linspace(-2*results[:, 2].std(), 2*results[:, 2].std(), N_bin)
    y = sigmoid(c, *popt)
    pylab.plot(c, y)
    pylab.text(50, 0.5, com)
    #pylab.text(0.05, 0.8, 'mean : %0.2f , slope : %0.2f ' %(np.mean(mean), np.mean(slope)))
    pylab.ylim(-.05, 1.05)
    #pylab.legend(loc='best')
```
