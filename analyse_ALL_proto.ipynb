```python
N_bin = 20
shift_bin = np.linspace(-20, 20, N_bin+1)
psy = np.ones((N_bin, 2, 9)) * .5
```

```python
def analyse(fig_list, param_Name):
    for i_bin in range(N_bin):
        for i_phi, phi in enumerate([1, -1]):
            ind_phi = results[:, 1] == phi
            for i_fig in range(1,10):
                ind_fig = results[ind_phi, 4] == i_fig
                ind_shift = shift_bin[i_bin] < results[ind_phi, 2][ind_fig]
                ind_shift *= results[ind_phi, 2][ind_fig] <= shift_bin[i_bin+1]
                psy[i_bin, i_phi, i_fig-1] = results[ind_phi, 0][ind_fig][ind_shift].mean()
            
    shift_middles = (shift_bin[:-1] + shift_bin[1:])/2.
    alpha = .8
    fig = plt.figure(figsize=(12,5))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)

    for i_fig, color in zip(fig_list, ['r', 'g', 'b']):
        ax1.plot(shift_middles, psy[:, 0, i_fig-1] * (.8 + .1*i_fig), c=color, alpha=alpha)
    ax1.vlines([0], ymin=-1, ymax=1)
    for i_fig, color, label in zip(fig_list, ['r', 'g', 'b'], param_Name):
        ax2.plot(shift_middles, psy[:, 1, i_fig-1] * (.8 + .1*i_fig), c=color, alpha=alpha, label=label)
    ax2.vlines([0], ymin=-1, ymax=1)
    for ax in [ax1, ax2]:
        ax.axis([-20, 20, -1.15, 1.15])
        _ = ax.set_xlabel('shift')
        _ = ax.set_ylabel('perceived direction')
    _ = ax2.set_ylabel('perceived direction')
    _ = ax2.legend(loc='right')
```

```python
exp1, exp2, exp3, exp4 = [1, 2, 3], [3, 4, 5], [3, 6, 7], [3, 8, 9]
param_Name1 = ['binary', 'natural', 'MC']
param_Name2 = ['B_theta = pi/32', 'pi/8', 'pi/2']
param_Name3 = ['theta = 0', 'pi/4', 'pi/2']
param_Name4 = ['B_sf = 0.1', '0.05', '0.5']
```

```python
analyse(exp1, param_Name1)
analyse(exp2, param_Name2)
analyse(exp3, param_Name3)
analyse(exp4, param_Name4)
```
